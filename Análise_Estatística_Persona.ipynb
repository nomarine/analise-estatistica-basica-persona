# Persona Analysis (3-5)

The objective of this analysis is to examine the patterns of the personas chosen to compose the Compendium of each game in the main franchise.

The given data considers:
- Only the main games, excluding spin-offs.
- Installments starting on Persona 3.
- The 'definitive' versions (Persona 3 Portable, Persona 4 Golden and Persona 5 Royal).
- Only personas that can be registered in the Compendium, meaning personas exclusive to Party Members or NPCs are excluded.
- DLC Personas (P5R only).

The reason for not taking into account Persona 1 and 2 is that the classification criteria by Arcanas in those games are directly tied to the Races of Shin Megami Tensei, a staple that is rarely observed in more recent personas where attributes are more relevant.

The data was obtained from the following source:
[Link to the data source](https://aqiu384.github.io/megaten-fusion-tool/home)

### Conclusion of the Persona Frequency Analysis by Arcana

**P3**

It is noticeable that the **distribution is nearly balanced**, **ranging from 7 to 9 Personas per Arcana**, with the **exception of Judgement and Aeon** arcanas, which contain the lowest number of Personas in the distribution.  
One can conclude that **Aeon and Judgement have fewer Personas because they are near-endgame Arcanas**, thus having higher-level Personas.  

**P4**

The **frequency is more variable** in this distribution, **ranging from 6 to 11**, having the **World Arcana as an outlier** since it contains only one Persona. **Temperance is the Arcana with most Personas**, having 11 in total. The **Arcanas with fewer Personas are, respectively, Lovers and Fortune with 6 and Judgement and Tower with 7**.  
**Once again Judgement is one of the Arcanas with fewer personas**, thus **corroborating** with assumption **that it contains the most high-leveled and powerful Personas** in the installment.  

**P5**

This **distirbution stands out** given the **Fool Arcana has a higher quantity of Personas not only on its own installment but also in comparison to the others**, having 16 in total. This can be explained given the DLC personas are the main personas of the protagonists from  previous installment, each having two variations, the standard and Picaro.  
Once again, the **outlier is the World Arcana**, with 2 personas only, those also from DLC with two variations.  
Excluding those Arcanas tainted by the DLC, **the Arcanas with most and fewer personas are, respectively, Moon and Hierophant.**  
It is interesting to notice that the **Judgement Arcana breaks its pattern not being one the the Arcanas with fewer Personas** in contrast to the previous installments.

**General**

It is perceivable that P3 and P4 maintain some pattern between them in comparison of P5 which presents a slight variation.  
Analysizing the level's mean and median by Arcana we can see that the Judgement contains the higher-level personas, corroborating with the assumption that it is tied to the quantity of Personas. But it's worth noting that the Tower doesn't follow this asusmption, seeing that it also contains high-level personas but only in P4 it is among the arcanas with fewer personas.
